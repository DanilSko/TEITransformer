```python
from tei_transformer import TEITransformer
import os
import numpy as np
from collections import defaultdict
```

```python
from timeit import default_timer as timer
```

```python
import sys
import warnings

if not sys.warnoptions:
    warnings.simplefilter("ignore")
```

## Data

```python
dir_path = "../../test/data"
schema_path = "../../test/schema.rng"
```

## Client interface initialization

```python
TT = TEITransformer(scenario='drama')
```

## Time profiling

```python
def feed_data(tei_path, schema_path, **kwargs):
    TT.load_tei(tei_path, schema_path=schema_path)
    start = timer()
    TT.transform(**kwargs)
    end = timer()
    run_time = end - start
    return run_time


def run_test(n=10, **kwargs):
    global dir_path, schema_path
    run_pereiods = defaultdict(list)
    for _ in range(n):
        for filename in os.listdir(dir_path):
            if not filename.startswith('.'):
                tei_path = os.path.join(dir_path, filename)
                run_time = feed_data(tei_path, schema_path, **kwargs)
                run_pereiods[filename].append(run_time)
    return run_pereiods

def print_run_time(run_pereiods):
    for file in run_pereiods:
        run_time = np.round(np.mean(run_pereiods[file]), 2)
        print("{}\t{}".format(file, run_time))
```

# Test1

Mean run time of files of different sizes.

### HTML

```python
run_pereiods_html = run_test(
    n=10, output_format='html', keep_all=True,
    full_page=True, enable_valid=False, custom_css_path=None
)
```

```python
print_run_time(run_pereiods_html)
```

## DOCX

```python
run_pereiods_docx = run_test(
    n=10, output_format='docx', keep_all=True,
    enable_valid=False, custom_css_path=None
)
```

```python
print_run_time(run_pereiods_docx)
```

## JSON

```python
run_pereiods_json = run_test(
    n=10, output_format='json', 
    enable_valid=False,
)
```

```python
print_run_time(run_pereiods_json)
```
